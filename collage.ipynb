```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
```

```python
l = []
top_left = cv2.imread("./jmp/top_left.jpg")
l.append(top_left)
top_right = cv2.imread("./jmp/top_right.jpg")
l.append(top_right)
bottom_left = cv2.imread("./jmp/bottom_left.jpg")
l.append(bottom_left)
bottom_right = cv2.imread("./jmp/bottom_right.jpg")
l.append(bottom_right)
for i in range(len(l)):
    l[i] = cv2.resize(l[i],(200,200))
    l[i] = cv2.cvtColor(l[i],cv2.COLOR_BGRA2RGBA)
```

```python
for img in l:
    plt.imshow(img)
    plt.show()
```

```python
add = [[2,1,2,1],[2,1,1,2],[1,2,2,1],[1,2,1,2]]
    
```

```python
for i in range(len(l)):
    l[i] = cv2.copyMakeBorder(l[i], 5*add[i][0],  5*add[i][1],  5*add[i][2],  5*add[i][3], cv2.BORDER_CONSTANT, None, value = 0)
```

```python
for i in range(len(l)):
    plt.imshow(l[i])
    plt.show()
```

```python
img = np.vstack((np.hstack((l[0],l[1])),np.hstack((l[2],l[3]))))
plt.imshow(img)
```
